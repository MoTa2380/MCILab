```python
import cv2
import matplotlib.pyplot as plt
import pytesseract
from preprocess import get_grayscale, increase_contrast, thresholding, remove_noise
```

```python
image_path = "/home/mohammad/Desktop/mcinext/llm/sandbox/ocr_pipeline/input_pdf/page_43.png"
image = cv2.imread(image_path)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(10,10))
plt.imshow(image_rgb)
plt.axis("off")
plt.show()
original = pytesseract.image_to_string(image_rgb, lang="fas")
print(original)
```

```python
image = get_grayscale(image=image)
plt.figure(figsize=(10,10))
plt.imshow(image, cmap="gray")
plt.axis("off")
plt.show()
gray = pytesseract.image_to_string(image=image, lang="fas")
print(gray)
```

```python
image = remove_noise(image=image)
plt.figure(figsize=(10,10))
plt.imshow(image, cmap="gray")
plt.axis("off")
plt.show()
noise = pytesseract.image_to_string(image=image, lang="fas")
print(noise)
```

```python
image = thresholding(image=image)
plt.figure(figsize=(10,10))
plt.imshow(image, cmap="gray")
plt.axis("off")
plt.show()
threshold = pytesseract.image_to_string(image=image, lang="fas")
print(threshold)
```
